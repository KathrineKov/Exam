Task 3
Create the following DataFrame:
student name
attendance of the DevEssentials course (in %)
mark (in [0..100] points)
Not too long, 5 lines are enough, to save time.
Draw a plot which shows a correlation (if any) between attendance and mark based on the given dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#DataFrame
data = {
    'student_name': ['Ilya', 'Kate', 'Feodosiy', 'Denis', 'Eva'],
    'attendance': [85, 92, 78, 96, 56],
    'mark': [88, 95, 70, 93, 60]
}

df = pd.DataFrame(data)
```

```python
#Plot
plt.figure(figsize=(8, 5))
plt.scatter(df['attendance'], df['mark'], color='red')
plt.title('Correlation between Attendance and Marks')
plt.xlabel('Attendance (%)')
plt.ylabel('Mark (out of 100)')
plt.grid(True)
plt.show()
```
